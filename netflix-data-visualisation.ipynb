### 3. DATA CLEANING

This is the most crucial phase of any Data Analysis project. Basically, remove null or irrelevant data and fill in values with respective means or medians. 

#### a) Removing Duplicates

**i) Checking for Duplicates**

```python
netdf.duplicated().sum()
```

Since, there are no duplicates, we will move on to missing values

#### b) Handling Missing Values

**i) Checking for missing values**

```python
netdf.isnull().sum()
```

It would definetely be better to drop the rows as we will still remain with plenty of data to spare

**ii) Drop missing rows**

```python
netdf.dropna(axis=0,how='any',inplace=True)
netdf1.dropna(axis=0,how='any',inplace=True)
```

```python
netdf.shape
```

```python
netdf1.shape
```

**c) Dropping irrelevant values**

The following attributes: "show_id" and "cast" will not produce any relevant visualisations with respect to any movies or tv shows produced by Netflix and hence, we must remove the column.

```python
netdf=netdf.drop("show_id",axis=1)
netdf=netdf.drop("director",axis=1)
netdf=netdf.drop("cast",axis=1)
netdf.shape
```

```python
netdf1=netdf1.drop("show_id",axis=1)
netdf1=netdf1.drop("cast",axis=1)
```

**d) Fixing values**

We can see that "country" and "listed_in" has multiple values however it is classified as one seperate value. We have to split the data into new rows.

```python
netdf["country"] = netdf["country"].str.split(',')
netdf["listed_in"]=netdf["listed_in"].str.split(',')
netdf= netdf.explode("country")
netdf= netdf.explode("listed_in")

netdf
```

Finally, we have gained the necessary numerical insights and cleaned the dataset. Let's visualize the data.

### 4. Data Visualisation

No data analysis is complete without pictorical representation of the data.

#### Histographs

**1. Types of Movies and TV Shows based on rating**

```python
plt.figure(figsize=(12,10))
plt.title("Rating of Movies and TV Shows")
plt.xlabel('Rating of Movies and TV Shows')
sns.histplot(x="rating",hue='type',data=netdf)
plt.show()
```

**2. Number of movies released on Netflix over the years**

```python
plt.figure(figsize = (12,8))
sns.histplot(x='release_year',data=netdf)
plt.title("Number of movies released over the years")
```

#### Bar graph

**1. Top 10 Genres streamed on Netflix**

```python
sns.countplot(y='listed_in', order=netdf['listed_in'].value_counts().index[0:10],data=netdf)
plt.title("Top 10 Genres streamed on Netflix")
```

**2. Top 10 Directors on netflix**

```python
sns.countplot(y='director', order=netdf1['director'].value_counts().index[0:10],data=netdf1)
plt.title("Top 10 Directors active on Netflix")
```

#### Pie Charts

**1. Ratio of TV Shows to Movies**

We use another dataframe netdf1 to retain the same number of movies and tv shows as in the original dataset since, we have split all the data into different rows in netdf

```python
type_c = list(netdf1['type'].value_counts())
color = ['red','green']
type_lis = list(netdf1['type'].value_counts().index)
fig = px.pie(values=type_c,names=type_lis)
fig.update_layout(title_text='Ratio of TV Shows to Movies', title_x=0.5)
fig.update_traces(rotation=90,textposition='inside',textinfo='label+percent+value',marker=dict(colors=color))
```

**2. Number of TV Shows based on Season**

```python
netdf[netdf['type']=="TV Show"]['duration'].value_counts().plot(kind="pie")
```

#### Scatter Plots

**1. Relationship Between Release Year and Rating**

```python
x=netdf1['release_year']
y=netdf1['rating']
plt.figure(figsize = (12,8))
plt.scatter(x, y)
plt.show()
```
